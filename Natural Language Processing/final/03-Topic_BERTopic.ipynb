```python
import pandas as pd 
import numpy as np
from bertopic import BERTopic
import re
```

```python
import nltk as nltk
import nltk.corpus  
from nltk.text import Text
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataPath = "/content/drive/MyDrive/NLP/"
```

```python
df = pd.read_csv(dataPath+'cleaned_df_topics.csv')
```

```python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
```

```python
df['cleaned_text'] = df['text'].apply(lambda x: re.sub("\n", " ", x))

# remove urls
#pattern = "^http[s]?:\/\/(www\.)?(.*)?\/?(.)*"
# pattern = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))'
pattern = "((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*"
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub(pattern, ' ', x))

# remove punctuations
# df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x))

# remove numbers
# df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub(r'[0-9]', '', x))

# keep only alphabetic words
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))

# to lower case
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: x.lower())
```

```python
df['text_without_stopwords'] = df['cleaned_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
```

```python
df.to_csv(dataPath+'cleaned_df_topics.csv', index=False)
```

```python
docs = df['text_without_stopwords']
```

```python
topic_model = BERTopic(verbose=True, nr_topics=50)
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.get_topic_info()
```

```python
topic_model.get_topic(6)
```

```python
topic_model.visualize_topics()
```

```python
topic_model.visualize_barchart(top_n_topics=50)
```

```python
import pickle
```

```python
pickle.dump(topic_model, open(dataPath+'topic_model.pkl', 'wb'))
```

## BERTopic on title

```python
df[['cleaned_title']].head(20)
```

```python
df['cleaned_title'] = df['title'].apply(lambda x: re.sub("\n", " ", x))
```

```python
df['cleaned_title'] = df['cleaned_title'].map(lambda x: re.sub('[^a-zA-Z0-9 ]', '', str(x)))

# to lower case
df['cleaned_title'] = df['cleaned_title'].apply(lambda x: x.lower())
```

```python
df['cleaned_title'] = df['cleaned_title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
```

```python
title = df['cleaned_title']
```

```python
topic_model = BERTopic(verbose=True, nr_topics=50)
topics, probs = topic_model.fit_transform(title)
```

```python
topic_model.get_topic_info()
```

```python
topic_model.visualize_topics()
```

```python
topic_model.visualize_barchart(top_n_topics=50)
```

```python
pickle.dump(topic_model, open(dataPath+'topic_model_title.pkl', 'wb'))
```
