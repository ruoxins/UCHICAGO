```python
import requests
import json
```

```python
response = requests.get("https://coronavirus.m.pipedream.net/", verify=False)
# print(response.status_code)
# print()
```

```python
json_response = response.json()
# print(json_response)
json_response = json_response['rawData'] # get only table data
dictionary = json.dumps(json_response, sort_keys = True, indent = 4)
# print(dictionary)
```

```python
import csv
import pandas as pd

df = pd.json_normalize(json_response)
df.head(10)
```

```python
# delete empty columns
df = df.drop(columns=['Recovered', 'Active'])

# delete 'FIPS', 'Admin2' columns - These are only available for regions in the US.
df = df.drop(columns=['FIPS', 'Admin2'])

# Rename the columns
df = df.rename(columns={"Lat": "Latitude", "Long_": "Longitude"})
df.head(10)
```
